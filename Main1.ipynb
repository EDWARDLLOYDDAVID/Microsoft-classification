```python
import pandas as pd



# Calculate the frequency of each combination
combination_counts = train_df.groupby(['CountryCode', 'State', 'City']).size().reset_index(name='count')

# Sort by 'count' in descending order and select the top 5
top_5 = combination_counts.sort_values(by='count', ascending=False).head(5)

# Create a mapping for the top 5 combinations
top_5_mapping = {tuple(row): i+1 for i, row in enumerate(top_5[['CountryCode', 'State', 'City']].values)}

# Function to encode based on top 5
def encode_row(row):
    key = (row['CountryCode'], row['State'], row['City'])
    return top_5_mapping.get(key, 0)  # Encode top 5 with 1-5, others with 0

# Apply the encoding function to the DataFrame
train_df['Feature1'] = train_df.apply(encode_row, axis=1)

print(train_df)
```
